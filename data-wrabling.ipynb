# Data Wrabling

**Import library's**

```python
import pandas as pd
import numpy as np
from datetime import datetime
```

**Load dataset and Make the date-time as an index**

```python
def parse(x):
    return datetime.strptime(x, "%Y %m %d %H")

dataset = pd.read_csv("pollution.csv", parse_dates=[["year", "month", "day",
                                                  "hour"]], index_col=0, date_parser=parse)
```

```python
dataset.head(5)
```

**Drop the No columns**

```python
dataset.drop("No", axis=1, inplace=True)
dataset.head(5)
```

**Change the column names and index**

```python
dataset.columns = ["pollution", "dew", "temp", "press", 
                   "wnd_dir", "wnd_spd", "snow", "rain"]
dataset.index.name = "date"
```

```python
dataset.head(5)
```

**Mark 0 to NA values in pollution column**

```python
dataset["pollution"].fillna(0, inplace=True)
dataset.head(5)
```

**Now drop the first 24 hours**

```python
dataset = dataset[24:]
dataset.head(5)
```
